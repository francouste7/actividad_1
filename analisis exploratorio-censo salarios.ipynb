# Analisis Outliers

```python
Q1 = censo['Age'].quantile(0.25) # percentil 25
Q3 = censo['Age'].quantile(0.75) # percentil 75
iqr = Q3 - Q1
```

```python
umbral_superior = Q3 + 1.5 * iqr
umbral_inferior = Q1 - 1.5 * iqr
print("la edad por debajo de : ",umbral_inferior, " es un outlier")
print("la edad por arriba de : ",umbral_superior, " es un outlier")
```

> OUTLIERS: tanto los graficos como la formula anterior indica que, a partir de los 78 empiezan los Outliers

```python
outliers = censo[(censo['Age'] > umbral_superior) | (censo['Age'] < umbral_inferior)]
print ("el feature Age cuenta con" , len(outliers) , "outliers")
```

> DIAGRAMA DE DISPERSION EN BASE A LOS OUTLIERS

```python
plt.scatter(censo.index,censo['Age'], label='datos', alpha=0.2)
plt.scatter(outliers.index,outliers['Age'], color='red', label='outliers')
plt.legend()
```
