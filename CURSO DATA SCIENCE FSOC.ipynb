```python
#funcion que calcule la media

notas = [4,5,6,7,8,2,4,5,7]
def media (notas):
    suma=0
    cantidad_notas = len(notas) #cantidad de notas del vector notas
    for nota in notas:
        suma += nota
    print ("mi promedio de notas es:" , suma/cantidad_notas)
    
```

# Estadistica
Es una rama de las matematicas que permite hacer varias cosas: recopilar,  organizar informacion y analizar datos segun lo que necesitemos.

Dos areas de la Estadistica:

--> DESCRIPTIVA: se dedica a la descripcion, visualizacion y resumen de datos originados a partir de fenomenos de estudio

--> INFERENCIAL: se dedica a la generacion de modelos y predicciones asociadas a fenomenos.

# Tipos de Datos

Cualitativos/categoricos: son las propiedades categoricas que pueden usarse para describir un objeto. Ejemplo: nacionalidad, sexo, tipo de fruta

Cuantitativos: representan una cantidad. Por ejemplo: edad, presion arterial, etc. 

# Tipos de Variables

-- Cualitativo nominal:son propiedades categoricas que permiten clasificar a la poblacion(objeto de estudios) en categorias excluyentes. 

--Cualitativo ordinal:los valores pueden ser ordenados de acuerdo a un criterio. Importa el orden. Ejemplo: encuesta, estado de un paciente

--Cuantitativo  Discreto: tiene un conjunto de valores finito. Ejemplo: cantidad de alumnos por curso

--Cuantitativo Continuo: pueden tomar todos los valores de un determinado intervalo. Ejemplo: altura, tiempo, volumen

> Outliers
- 
Valores atípico- 

Rango intercuartil = Q3 -Q- 1

Umbral superior = DATO > Q3 + 15- IQR

Umbral inferior = Dato < Q1 - 1.5IQR

```python
Q1 = st['bmi'].quantile(0.25) # percentil 25
Q3 = st['bmi'].quantile(0.75)
iqr = Q3 - Q1
```

```python
umbral_superior = Q3 + 1.5 * iqr
umbral_inferior = Q1 - 1.5 * iqr
print("Un valor de BMI por debajo de : ",umbral_inferior, " es un outlier")
print("Un valor de BMI por arriba de : ",umbral_superior, " es un outlier")
```

```python
# Agarrar de la columna bmi los valores mayores a mi umbral superior o los menores a mi umbral inferior
outliers = st[(st['bmi'] > umbral_superior) | (st['bmi'] < umbral_inferior)]
```

```python
len(outliers)
```

DIAGRAMA DE DISPERSION EN BASE A LOS OUTLIERS

```python
plt.scatter(st.index,st['bmi'], label='datos', alpha=0.2)
plt.scatter(outliers.index,outliers['bmi'], color='red', label='outliers')
plt.legend()
```
